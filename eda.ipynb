```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fishRecognition_GT/clusters.csv')
df.head()
```

```python
df = df.groupby("species").size().reset_index(name='count').sort_values('count', ascending=False)
df
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['species'], df['count'])
plt.title('Species Count - Bar Plot')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['species'], df['count'])
plt.yscale('log')
plt.title('Species Count - Log Scale Bar Plot')
plt.xlabel('Species')
plt.ylabel('Count (Log Scale)')
plt.xticks(rotation=90)
plt.show()
```
